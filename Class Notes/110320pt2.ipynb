Using the following data,

```r
# Load the Tidyverse

library('tidyverse')
```

```r
# From http://stackoverflow.com/questions/1181060
stocks = tibble(
  time = as.Date('2009-01-01') + 0:999,
  X = rnorm(1000, 0.35, 1),
  Y = rnorm(1000, 0, 2),
  Z = rnorm(1000, 2, 4)
)

stocks %>% print
```

Re-create the following figure:

(*Hints: use* `pivot_longer` *to re-shape the data.*)

<img src="figs/stock_gather.png" style="width: 500px;" />

```r
stocks_longer = stocks %>%
    pivot_longer(cols = c(X,Y,Z), names_to = 'Stock', values_to = 'Price')
head(stocks_longer)
```

```r
ggplot(data = stocks_longer, aes(x = Stock, y = Price)) + 
  geom_point(position = position_jitter(h = 0, w = 0.3), alpha = 0.3)
```
