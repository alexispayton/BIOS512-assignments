```r
library(tidyverse)
measles_df = read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-25/measles.csv")
```

Make a bar chart of the frequency of values for one "categorical" column in the data you chose.

```r
ggplot(data = measles_df, aes(y = state)) + 
  geom_bar()
```

Using `forcats` functions, plot an ordered example of your bar chart from above. Don't forget to "lump" low frequency values.

```r
sorted_measles_df = measles_df %>%
    mutate(state = state %>% fct_infreq())  #increasing frequency

ggplot(data = sorted_measles_df, aes(y = state)) + 
  geom_bar()
```

```r
sorted_measles_df2 = measles_df %>%
    mutate(state = state %>% fct_infreq() %>% fct_rev())  

ggplot(data = sorted_measles_df2, aes(y = state)) + 
  geom_bar()
```

```r
#can also pull out state from df to factor and reassign col in original df
measles_df$state %>% fct_infreq() #factor is based on increasesing frequency of the state
```

```r
sorted_measles_df3 = measles_df %>%
    mutate(state = state %>% fct_lump(n = 10) %>% fct_infreq() %>% fct_rev() )
ggplot(data = sorted_measles_df3, aes(y = state)) + 
  geom_bar()
```
