Import the "Avengers" data from the [FiveThirtyEight data repository on GitHub](https://github.com/fivethirtyeight/data/blob/master/avengers/avengers.csv).

```r
library(tidyverse)
avengers = read_csv("https://github.com/fivethirtyeight/data/raw/master/avengers/avengers.csv")
```

Make an ECDF plot of `Appearances`.

```r
ggplot(data = avengers, aes(x = Appearances)) +
  stat_ecdf(geom = "point", size = 2)

#median number of appearances = x value where y = 0.5
```
