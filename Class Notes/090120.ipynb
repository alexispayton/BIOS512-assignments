5. Create a facets plot where each facet represents one iris species. 
(You may have to resize the output again.)

```r
options(repr.plot.width = 10, repr.plot.height = 3, repr.plot.res = 300)
ggplot(data = iris, aes(x = Petal.Width, y = Petal.Length, color = Species)) + 
  geom_point(alpha = 0.7) + #alpha makes points transluscent 
  facet_wrap(~Species)
```
