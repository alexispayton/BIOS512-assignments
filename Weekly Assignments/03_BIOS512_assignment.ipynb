````r
options(repr.plot.width = 8, repr.plot.height = 3)

ggplot(data = penguins_long, aes(x = species, y = value, color = sex)) + 
  geom_point(position = position_jitter(h = 0, w = 0.15), alpha = 0.5) + 
  stat_summary(fun = median, 
                     fun.min = median, 
                     fun.max = median, 
                     geom = "crossbar", 
                     width = 0.75, 
                     size = 0.2,
                     mapping = aes(group = species)) + #adds median to chart
  facet_wrap(~name, ncol = 4, scales = 'free_y') + 
  theme(axis.text.x = element_text(hjust = 1, angle = 45))
````
