````r
df_marathon = read.csv('https://raw.githubusercontent.com/chuckpr/BIOS512/main/in-class-exercises/data/marathon.csv')

df_filtered = df_marathon %>% 
    filter(country %in% c('United States', 'Kenya', 'Ethiopia', 'Japan'))

df_country = df_marathon %>%
    filter(country %in% c('United States', 'Kenya', 'Ethiopia', 'Japan')) %>%
    group_by(country, MF) %>%
    summarize(race.time = mean(race.time), year = mean(year))
````

````r
options(repr.plot.width = 5, repr.plot.height = 5, repr.plot.res = 300)
ggplot(df_marathon, aes(x = year, y = race.time)) + 
  geom_point(color = 'gray50') + 
  geom_point(data = df_filtered, aes(color = country)) + 
  facet_wrap(~MF, scales = "free", ncol = 1) + 
  geom_text_repel(
  segment.color = 'transparent',
    nudge_y = 0.2,
    data = df_country, 
    aes(label = country, color = country)) +
  labs(title = 'Boston marathon winners are getting faster and faster', y = 'Winning Race Time(hrs)',
       subtitle = "Winning race times for the men's and women's categories") +
  #ggtitle('Boston marathon winners are getting faster and faster') + #plot title
  theme(legend.position = 'none', plot.background = element_rect(fill = 'black'), 
        axis.title.y = element_text(color = "gray50", margin = margin(r = 12)), #changing y axis title
        plot.title = element_text(color = "gray50", hjust = 1.5), #changing plot title
        plot.subtitle = element_text(color = "gray50", hjust = -0.5, size = 8), #changing plot subtitle
        plot.margin = unit(c(0.5,0.5,0.5,0.5), "cm"),
        panel.grid.minor = element_blank(), #removed minor grid lines
        panel.grid = element_line(color = "gray50"), #making grid lines gray
        panel.background = element_rect(fill = 'black'),#entire background black
        strip.text.x = element_text(color = "gray50"), #changing color of facet titles
        strip.background = element_rect('black'))  #making background of facet titles black
````
